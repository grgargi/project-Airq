# Air Quality Analysis Project 

This project analyzes air quality levels across various locations and times, aiming to identify significant patterns and assess the potential health impacts of poor air quality. We examine the data, perform thorough preprocessing, and explore key trends to build a foundation for predictive modeling.

Data Understanding

The dataset consists of various features related to air quality, including measurements of pollutant levels, geographic indicators, and time-based information.
Our target variable, Data Value, represents the air quality level we aim to predict, allowing for assessments of patterns over time and across regions.

Data Preprocessing

To ensure data quality and prepare it for analysis, we performed the following preprocessing steps:

Handling Missing Values: Missing values were treated by using imputation methods for critical features, such as the median for numerical columns with moderate missing values, and dropping columns with excessive null values that offered minimal impact on analysis.

Outlier Detection and Removal: Outliers were identified using the Interquartile Range (IQR) method, which helped remove extreme values that could skew the analysis. Boxplots were used to visualize and verify the effect of outlier removal.
Data Encoding and Scaling: Categorical variables were encoded, and numerical features were scaled using StandardScaler to ensure that all features contribute proportionately to the analysis.

Exploratory Data Analysis (EDA)

We conducted a comprehensive EDA to uncover trends, relationships, and insights within the data:

Correlation Analysis: A correlation heatmap revealed relationships between features, helping to identify which variables might influence air quality the most.
Pair Plots: Pair plots allowed us to explore relationships between numerical features, offering insight into trends and distributions.

Trend Analysis: Scatter plots and histograms illustrated how Data Value varied over time and across regions, providing an initial understanding of seasonal and geographic trends in air quality.
These initial analyses highlighted significant factors and patterns in air quality, setting the stage for predictive modeling.

### Handling Outliers
Outliers can distort the predictions of our model, so we use the Interquartile Range (IQR) method to detect and remove extreme outliers from the Data Value field. This process helps make the data more stable and reliable for model training

```python
# Detect outliers using IQR
Q1 = df['Data Value'].quantile(0.25)
Q3 = df['Data Value'].quantile(0.75)
IQR = Q3 - Q1

# Filter data without extreme outliers
df = df[~((df['Data Value'] < (Q1 - 1.5 * IQR)) | (df['Data Value'] > (Q3 + 1.5 * IQR)))]
print("Shape after outlier removal:", df.shape)
```

```python
# Boxplot before removing outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Data Value'])
plt.title("Boxplot of 'Data Value' Before Outlier Removal")
plt.show()

# Outlier removal using IQR
Q1 = df['Data Value'].quantile(0.25)
Q3 = df['Data Value'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['Data Value'] < (Q1 - 1.5 * IQR)) | (df['Data Value'] > (Q3 + 1.5 * IQR)))]

# Boxplot after removing outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Data Value'])
plt.title("Boxplot of 'Data Value' After Outlier Removal")
plt.show()
```

#### Encoding and Scaling Features
Before modeling, ensure that categorical features are encoded and numerical features are scaled.  A preprocessing pipeline to handle both.

```python
target = 'Data Value'
X = df.drop(columns=[target])
y = df[target]

# Identify categorical and numerical columns
categorical_features = X.select_dtypes(include=['object']).columns.tolist()
numerical_features = X.select_dtypes(include=['float64', 'int64']).columns
```

#### Preprocessing Pipeline

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer


# Preprocessing for numerical and categorical data
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

# Combine transformations
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```
